```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("retail_transacation_2000.csv")

# Basic info
df.head()
df.info()
df.describe()

# Check missing values
df.isnull().sum()

# Remove duplicates
df = df.drop_duplicates()

# Create TotalAmount column
df["TotalAmount"] = df["Quantity"] * df["Price"]

# Top 10 products by sales
top_products = df.groupby("Product")["TotalAmount"].sum().sort_values(ascending=False).head(10)

plt.figure()
top_products.plot(kind="bar")
plt.title("Top 10 Products by Sales")
plt.show()

# Save this figure
plt.figure()
top_products.plot(kind="bar")
plt.title("Top 10 Products by Sales")
plt.savefig("top_products.png")
plt.close()

# Price distribution
plt.figure()
sns.histplot(df["Price"], bins=20)
plt.title("Price Distribution")
plt.savefig("price_distribution.png")
plt.close()

# If Date column exists
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"])
    daily_sales = df.groupby("Date")["TotalAmount"].sum()

    plt.figure()
    daily_sales.plot()
    plt.title("Sales Trend Over Time")
    plt.savefig("sales_trend.png")
    plt.close()

print("EDA Completed. Images saved.")
```
